```
# Breadth-first search (BFS) algorithm
def bfs(graph, initial):
    visited = []
    queue = [initial]

    while queue:
        node = queue.pop(0)
        if node not in visited:
            visited.append(node)
            neighbours = graph[node]
            
            for neighbour in neighbours:
                queue.append(neighbour)
    return visited
 

# Initializing a graph
# Sibiu, Fagaras, Rimnicu, Pitesti, Craiova, Bucharest, Giurgiu
graph = {}

graph['Sibiu'] = ['Fagaras', 'Rimnicu']
graph['Fagaras'] = ['Sibiu','Bucharest']
graph['Rimnicu'] = ['Sibiu', 'Pitesti', 'Craiova']
graph['Pitesti'] = ['Rimnicu','Craiova', 'Bucharest']
graph['Craiova'] = ['Rimnicu', 'Pitesti']
graph['Bucharest'] = ['Giurgiu','Pitesti', 'Fagaras']
graph['Giurgiu'] = ['Bucharest']


print(bfs(graph, 'Sibiu'))
```
